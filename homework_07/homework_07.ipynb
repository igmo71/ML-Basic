### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
data_2 = np.arange(10, 50)
print(data_2, end="\n\n")

data_2 = data_2[::-1]
print(data_2, end="\n\n")

data_2 = data_2[1::2]
print(data_2)
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
data_4 = np.random.rand(4, 3, 2)
print(data_4, end="\n\n")

print(f'min = {data_4.min()}')
print(f'max = {data_4.max()}')
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
data_5 = np.arange(1, 25).reshape(6, 4)
print(data_5, end='\n\n')

data_6 = np.arange(2, 14).reshape(4, 3)
print(data_6, end='\n\n')

data_7 = np.dot(data_5, data_6)
print(data_7)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
np.random.seed(42)

data_8 = np.random.rand(7, 7) * 100
np.set_printoptions(precision=2, suppress=True)
print(data_8, end="\n\n")

print(f'mean = {np.mean(data_8):.2f}', end="\n\n")

print(f'std = {np.std(data_8):.2f}', end="\n\n")

data_8_normalized = (data_8 - np.mean(data_8)) / np.std(data_8)
print(data_8_normalized, end="\n\n")

print(f'normalized mean = {np.mean(data_8_normalized):.2f}', end="\n\n")

print(f'normalized std = {np.std(data_8_normalized):.2f}', end="\n\n")
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
plt.figure(figsize=(10, 6))
plt.title('Распределение Tip в зависимости от Time')
plt.xticks(range(0, 15, 1)) 
plt.yticks(range(0, 25, 2)) 
plt.grid(linestyle='--', alpha=0.3)
sns.histplot(data=tips, x='tip', hue='time', kde=True, bins=60)
plt.show()
```

```python
lunch = tips[tips['time'] == 'Lunch']['tip']
dinner = tips[tips['time'] == 'Dinner']['tip']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

sns.histplot(data=lunch, bins=30,  kde=True, ax=ax1)
ax1.set_title('Lunch')
ax1.grid(linestyle='--', alpha=0.3)

sns.histplot(data=dinner, bins=30,kde=True,ax=ax2, color='orange')
ax2.set_title('Dinner')
ax2.grid(linestyle='--', alpha=0.3)

fig.suptitle('Распределение Tip в зависимости от Time', fontsize=16)
plt.tight_layout()
plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
male = tips[tips['sex'] == 'Male']
female = tips[tips['sex'] == 'Female']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

sns.scatterplot(data=male, x='total_bill', y='tip', hue='smoker', ax=ax1)
ax1.set_title('Male')
ax1.grid(linestyle='--', alpha=0.3)

sns.scatterplot(data=female, x='total_bill', y='tip', hue='smoker', ax=ax2,)
ax2.set_title('Female')
ax2.grid(linestyle='--', alpha=0.3)

fig.suptitle('Взимосвязь между Total Bill и Tip для Male и Female в зависимости от Smoker', fontsize=16)
plt.tight_layout()
plt.show()
```
